```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
%matplotlib
#muestra la figura en una ventana aparte

fig = plt.figure()
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")
```

```python
#Line plots y scatter plots
#Líneas y líneas punteadas
%matplotlib inline

fig2 = plt.figure()
ax = plt.axes()

plt.plot(x, np.sin(x), linestyle="dashed")
plt.plot(x, np.cos(x))
```
